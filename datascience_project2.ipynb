```python
#Data Science Project2
#gapminder data for specific any country and its population

#import all useful library and gapminder dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mind=pd.read_csv('/content/drive/MyDrive/Python_File/Gap minder data.csv')
print(mind)
```

```python
#Give value of year in variable x
x = [i*1 for i in range(1952,2010,5)]
print("Year=",x)
```

```python
#in csv file country column convert to list
# this list is converted into set to get unique value of country
# This set is again converted into list to get unique value and in "country"
country = list(set(mind["country"].tolist()))

#Country is sorted to get alphabetical order
con = list(np.sort(country))
print("country in csv file:",con)
```

```python
#pop column converted to list
population = mind["pop"].tolist()
print(population)
```

```python
#Create 2D empty array of size 142*12
pop = np.zeros(shape=(142,12))

#Store the population record in this 2D array
i=0
for j in range(0,142):
  for k in range(0,12):
    pop[j][k] = round(population[i])
    i=i+1
print(pop)
```

```python
# This code collect data of popution of 4 country(You can choose any country in csv file)
ind_pop=pop[con.index('India')]
pak_pop=pop[con.index('Pakistan')]
cn_pop=pop[con.index('China')]
us_pop=pop[con.index('United States')]
print("India Population\n",ind_pop)
print("\nPakistan Population\n",pak_pop)
print("\nChina Population\n",cn_pop)
print("\nU.S Population\n",us_pop)
```

```python

#plot this 4 country data
plt.plot(x,ind_pop,label='India',color='blue')
plt.plot(x,pak_pop,label="Pakistan",color='red')
plt.plot(x,cn_pop,label="China",color='yellow')
plt.plot(x,us_pop,label="U.S.",color='green')

#Give x & y label in graph and show legend
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()

#plot the graph
plt.show()
```

```python
#gdpPercap conterted into list
gdp = mind["gdpPercap"].tolist()
print(gdp)
```

```python
#Store gdp data in 2D array
GDP = np.zeros(shape=(142,12))
i=0
for j in range(0,142):
  for k in range(0,12):
    GDP[j][k] = round(gdp[i])
    i=i+1
print(GDP)
```

```python
# This code collect data of gdp per capital of 4 country(You can choose any country in csv file)
ind_gdp=GDP[con.index('India')]
pak_gdp=GDP[con.index('Pakistan')]
cn_gdp=GDP[con.index('China')]
us_gdp=GDP[con.index('United States')]
print("India gdp\n",ind_pop)
print("\nPakistan gdp\n",pak_pop)
print("\nChina gdp\n",cn_pop)
print("\nU.S gdp\n",us_pop)
```

```python
#Plot 4 country's gdp data
plt.plot(x,ind_gdp,label='India',color='blue')
plt.plot(x,pak_gdp,label="Pakistan",color='red')
plt.plot(x,cn_gdp,label="China",color='yellow')
plt.plot(x,us_gdp,label="U.S.",color='green')

#Add lebel and legend and plot graph
plt.xlabel("Year")
plt.ylabel("GDP_Per_Capita")
plt.legend()
plt.show() 
```

```python
#Average Life of people in country's data converted to list
life = mind["lifeExp"].tolist()
print(life)
```

```python
#Store life list into 2D array 
Life = np.zeros(shape=(142,12))
i=0
for j in range(0,142):
  for k in range(0,12):
    Life[j][k] = round(life[i])
    i=i+1
print(Life)
```

```python
#This code collect data of Average life of 4 country(You can choose any country from csv file)
ind_life=Life[con.index('India')]
pak_life=Life[con.index('Pakistan')]
cn_life=Life[con.index('China')]
us_life=Life[con.index('United States')]
print("India Average Life\n",ind_pop)
print("\nPakistan Average Life\n",pak_pop)
print("\nChina Average Life\n",cn_pop)
print("\nU.S Average Life\n",us_pop)
```

```python
#Plot 4 country's data
plt.plot(x,ind_life,label='India',color='blue')
plt.plot(x,pak_life,label="Pakistan",color='red')
plt.plot(x,cn_life,label="China",color='yellow')
plt.plot(x,us_life,label="U.S.",color='green')

#Add legend, lebel and plot graph
plt.xlabel("Year")
plt.ylabel("Average Age")
plt.legend()
plt.show()
```
